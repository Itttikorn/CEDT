```python
positive_sentences = [
    "วันนี้เป็นวันที่ดี",
    "ฉันรู้สึกมีความสุข",
    "ชีวิตเต็มไปด้วยความหวัง",
    "ฉันรักครอบครัวของฉัน",
    "ฉันมีเพื่อนที่ดี",
    "สุขภาพแข็งแรง",
    "พอใจกับงานของฉัน",
    "ชีวิตมีความสุข",
    "พอใจกับสิ่งที่มี",
    "สนุกกับการเรียนรู้สิ่งใหม่ๆ",
    "ยินดีที่ได้ช่วยเหลือผู้อื่น",
    "สนุกกับการเดินทาง",
    "เพลิดเพลินกับการออกกำลังกาย",
    "ชอบอ่านหนังสือ",
    "ฟังเพลงอย่างมีความสุข",
    "สนุกกับการทำอาหาร",
    "เล่นกับสัตว์เลี้ยงอย่างสนุกสนาน",
    "ใช้เวลากับเพื่อนอย่างมีความสุข",
    "ใช้เวลากับครอบครัวอย่างมีความสุข",
    "พักผ่อนอย่างเต็มที่"
]

negative_sentences = [
    "วันนี้เป็นวันที่แย่",
    "ฉันรู้สึกเศร้า",
    "ชีวิตเต็มไปด้วยความทุกข์",
    "ฉันไม่ชอบงานของฉัน",
    "ฉันรู้สึกโดดเดี่ยว",
    "ไม่มีความสุขในชีวิต",
    "ความเครียดเข้ามาในชีวิต",
    "เหนื่อยล้าจากทุกสิ่ง",
    "ท้อแท้กับสถานการณ์",
    "ผิดหวังในตัวเอง",
    "ความโกรธเข้าครอบงำ",
    "หดหู่ใจ",
    "วิตกกังวลตลอดเวลา",
    "ไม่สบายใจในสิ่งที่เกิดขึ้น",
    "ไม่พอใจกับผลลัพธ์",
    "รู้สึกไม่มั่นคง",
    "ไม่ปลอดภัยในสภาพแวดล้อม",
    "สุขภาพไม่ดี",
    "รู้สึกแย่กับตัวเอง",
    "ไม่ชอบใจในสิ่งที่เห็น"
]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from pythainlp.tokenize import word_tokenize

tokenized_positive_sentences = [' '.join(word_tokenize(sentence)) for sentence in positive_sentences]
tokenized_negative_sentences = [' '.join(word_tokenize(sentence)) for sentence in negative_sentences]

pipeline = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('classifier', MultinomialNB())
])

data = tokenized_positive_sentences + tokenized_negative_sentences
labels = ['positive'] * len(positive_sentences) + ['negative'] * len(negative_sentences)

pipeline.fit(data, labels)
```

```python
import pandas as pd

test_cases = [
    "ฉันมีความสุขมากในวันนี้", 
    "ฉันรู้สึกเศร้าและโดดเดี่ยว", 
    "วันนี้เป็นวันที่ดีและสนุกสนาน",  
    "ฉันไม่พอใจกับผลลัพธ์ที่ได้",  
    "ยินดีต้อนรับเข้าสู้บ้านของเรา"  
]

expected_sentiments = ['positive', 'negative', 'positive', 'negative', 'positive']


predictions = pipeline.predict(test_cases)
results_df = pd.DataFrame({
    'Sentence': test_cases,
    'Expected Sentiment': expected_sentiments,
    'Predicted Sentiment': predictions,
    'Result': ['correct' if expected == prediction else 'incorrect' for expected, prediction in zip(expected_sentiments, predictions)]
})

results_df
```
